```python
import pandas as pd
from neuralprophet import NeuralProphet
from matplotlib import pyplot as plt
import pickle
```

```python
m = NeuralProphet()
```

```python
data = pd.read_csv("../Khammam.csv")
print(len(data.ds))
```

```python
print(len(data.ds.unique()))
```

```python
data.columns
```

```python
data = data[['ds', 'y']]
```

```python
data.head()
```

```python
print(len(data.ds.unique()))
print(len(data.ds))
```

```python
model = m.fit(data, freq='D', epochs=1000)
```

```python
future = m.make_future_dataframe(data, periods=387)
forecast = m.predict(future)
forecast.head()
```

```python
khammam_forecast = forecast[['ds', 'yhat1']]
khammam_forecast = khammam_forecast[22:]
```
