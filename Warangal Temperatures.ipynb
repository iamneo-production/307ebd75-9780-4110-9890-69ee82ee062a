```python
import pandas as pd
```

```python
df = pd.read_csv(r"..\full.csv")
df.head()
```

```python
df.dtypes
```

```python
from matplotlib import pyplot as plt
df.columns
```

```python
df1 = df.query('Mandal == "Warangal" and location == "Kasibugga"')
df1
```

```python
df2 = df.query('Mandal == "Warangal" and location.isnull()')
df2
```

```python
frames = [df1, df2]
df = pd.concat(frames)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
plt.plot(df['Date'], df['temp_max'], 'ro')
plt.show()
```

```python
data = df[['Date', 'temp_max']]
data
```

```python
data.dropna(inplace=True)
data.columns = ['ds', 'y'] 
data.tail()
```

```python
data = data.drop_duplicates()
```

```python
data.to_csv('Warangal.csv')
```
